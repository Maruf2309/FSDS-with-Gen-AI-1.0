#### Continue Statement

```python
for i in range(1, 10):
    if i == 5:
        continue  # contiue match 5 skip kore then continue korbe, get if ignore & continue
    print(i)
```

**Application**
- when we worked with csv / excel file have some corrupted / bad lines

```python
# Additional Examples of continue in loops
for i in range(1, 11):
    if i % 2 == 0:   # Skip even numbers
        continue
    print(i)
```

```python
# Example 2: Skip names starting with A
names = ["Alice", "Bob", "Alex", "John"]
for name in names:
    if name.startswith("A"):
        continue
    print(name)
```

```python
# Example 3: Skip negative numbers
numbers = [10, -5, 7, -3, 12]
for n in numbers:
    if n < 0:
        continue
    print(n)
```

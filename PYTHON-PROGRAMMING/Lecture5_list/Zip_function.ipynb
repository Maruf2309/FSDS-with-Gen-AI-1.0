# Zip Function

The Zip() function

```python
L1 = [1, 2, 3, 5]
L2 = [5, 6, 7, 8]
```

```python
zip(L1,L2) # its always return object
```

```python
list(zip(L1,L2))
```

When lenght is different

```python
L1 = [1, 2, 3, 5]
L2 = [5, 6, 7]
```

```python
zip(L1,L2)
```

```python
list(zip(L1,L2))
```

```python
roll = [1, 2, 3, 5]
student = ['Maruf','Alex','Bob','Shan']

list(zip(roll, student))
```

```python
# Using Loop
roll = [1, 2, 3, 5]
student = ['Maruf', 'Alex', 'Bob', 'Shan']

result = []
for i in range(len(roll)):
    result.append((roll[i], student[i]))

print(result)
```

```python
roll = [1, 2, 3, 5]
student = ['Maruf','Alex','Bob','Shan']

tuple(zip(roll, student))
```

```python
L1 = [1, 2, 3, 4]
L2 = [5, 6, 7, 8]

[i-j for i,j in zip(L2, L1)]
```

```python
L1 = [1, 2, 3, 4]
L2 = [5, 6, 7, 8]
L3 = [9,10]

list(zip(L1,L2,L3))
```

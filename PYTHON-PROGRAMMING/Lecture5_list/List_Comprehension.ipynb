# List Comprehension

List Comprehension provides a concise way of creating lists

Advantage of List Comprehension

```python
# Add 1 to 10 number to a list

L = []

for i in range(1,11):
    L.append(i)

print(L)
```

```python
# List Compresion to do the same
lc = [i for i in range (1,11)]  # list comprehion used for simple not for complex code, the var used for store, need to give in first
print(lc)
```

```python
# Scalar mulitiplication on a vector

v = [2, 3, 4]
s = -3

L = []
for i in v:
    result = i*s
    L.append(result)
print(L)

```

```python
lc = [ i*s for i in v]
print(lc)
```

# Conditional Statement

```python
# Print all the numbers divible by 5 in the range of 1 to 50

for i in range(1, 51):
    if i%5 == 0:
        print(i)
```

```python
lc = [ i for i in range (1,51) if i%5==0]  # Condition writes later/last
print(lc)
```

```python
# find languate which start with letter P

languages = ['java','python', 'php','ruby']

lc = [language for language in languages if language.startswith('p')]
print(lc)
```

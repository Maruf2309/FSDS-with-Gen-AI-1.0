```python
# List
l = [1, 2, 3]

# Memory Location
print("Memory Location of L :",id(l))

# List has different values
# Now show memory (all are different memory location)

print(id(l[0]))  #Location of 1
print(id(l[1]))  #Location of 2
print(id(l[2]))  #Location of 3
```

# Creating List

```python
# Empty
print([])
# 1D
print([1,2,3,4])
# 2D
print([1,2,3,4,[4,5]])
# 3D
print([[1,2],[3,4]], [[5,6],[7,8]])
print(list("hello"))
```

# Accesing Item from List

- Indexing
- Slicing

```python
l = [1,2,3,4,5]

print(l[2])     # Access 3
print(l[-1])
```

```python
# Multidimetional
l2 = [1,2,3,[4,5]]

print(l2[3][1])  # Accessing last 5
print(l2[-1][-1])
```

```python
# Extend
l1 = [1,2,3]
l2 = [6,7,9]
l1.extend(l2)
print(l1)
```

```python
l = [11,12,13]
l.extend("Hello")
print(l)
```

```python
l = [11,12,13]
l.append("Hello")
print(l)
```

```python
# Insert
l = [1,2,3,4]
l.insert(1,200)
print(l)
```

```python
# Deletion
l = [1,2,3,4]
del l[1:3]
print(l)
```

```python
# Remove
l = [1,2,3,4]
l.remove(4)
print(l)
```

```python
# Pop
l = [1,2,3,4]
l.pop()  # Opposive of append
l.pop()  # Second Pop
print(l)
```

```python
# Clear
l = [1,2,3,4]
l.clear()
print(l)
```

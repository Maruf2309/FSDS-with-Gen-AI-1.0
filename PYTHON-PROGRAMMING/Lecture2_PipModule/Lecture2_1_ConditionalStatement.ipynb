## Conditional Statement in Python

```python
# Loging Funcation demo - website

"""
Actual Credential :
email: mb@gmail.com
password :Mb123@122

"""

email = input("Enter your email: ")
password = input("Enter your password: ")

# Check Condition

actual_email = 'mb@gmail.com'
actual_password  = 'Mb123@122'

# Need  to check two conditions
if email == actual_email and password == actual_password:
    print("Login Successfully!")
else:
    print("Invalid Password!")

```

**Advance Level**
- 3 Conditions

```python
email = input("Enter your email: ")
password = input("Enter your password: ")

# Check Condition

actual_email = 'mb@gmail.com'
actual_password  = 'Mb123@122'

# Need  to check two conditions
if email == actual_email and password == actual_password:
    print("Login Successfully!")
elif email == actual_email and password != actual_password:
    print("Incorrect Password!")
else:
    print("Invalid Credentials!")
```

# Nested Conditions

```python
email = input("Enter your email: ")
password = input("Enter your password: ")

# Check Condition

actual_email = 'mb@gmail.com'
actual_password  = 'Mb123@122'

# Need  to check two conditions
if email == actual_email and password == actual_password:
    print("Login Successfully!")

elif email == actual_email and password != actual_password:
    print("please try again!")
    password = input("Enter your password again: ")

    if password == actual_password:
        print("Success!")
    else:
        print("Wrong!")

else:
    print("Invalid Credentials!")
```

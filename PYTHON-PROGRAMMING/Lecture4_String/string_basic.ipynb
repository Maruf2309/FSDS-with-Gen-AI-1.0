#### Access String

```python
s = "Welcome to Python"

# Access W from this strin

# Positive Index (Left to Right) start from 0 index

print(s[0])  # Object then give index Number
print(s[11]) # Positive
print(s[-6]) # Negaive
# print(s[50]) # Will get error!Strin index out of range

print(len(s))  # len is fun

## Negative Indexing  (Right to Left), index start from - 1
t = "Welcome to Python"

print('Negative Index: ',t[-1])  # Access last n from the string
print('Negative Index: ',t[-11]) # Access e(Last of Welecome) from the strign
 
```

### String Slicing

```python
s = "Welcome to Python"

# Accessing Welcome from s
# Lower bound & Upper bound, : sing
print(s[0:6]) # like range, last number excluded
print(s[0:7])
print(s[11:17]) # Access Python
print(s[11:50]) # No Error, because its start from starting from start index, but when get last, will ignore last limit like 50
print(s[11:60]) 
print(s[11:])   # without bound,then python conisdered to last
print("Welcome:",s[:]) 
print(s)
print(s[0:7:2])  # Step or Jump Params, Jump param always positive

# Negative Slicing

print(s[6::-1])  # Reversing Welcome
print(s[6::-2])  # Reversing Welcome
print(s[::-1])  # Full Reverse
print(s[::-2])  # Full Reverse with 1 Jump (2 means 1 character, 3 means 2 character(as last excluded))
```

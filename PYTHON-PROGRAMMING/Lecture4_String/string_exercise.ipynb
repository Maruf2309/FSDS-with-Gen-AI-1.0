**2.Write a program that can check whether a given string is palindrome or not**

- abba
- mam
- madam

```python
# Check palindrom (both side writting same like abba)

word = input("Enter your word: ")

flag = True

for i in range(0, len(word)//2):    # mod becuse to make comparision, say input will be "abba"
    if word[i] != word[len(word) - i - 1]:
        flag = False
        print("Not a Palindrome")
        break

if flag:          # or if flag
    print("Palindrome")
```

- Count

```python
s = "My name is Maruf"
```

```python
print(s.count("M"))
print(s.count("U"))
print(s.count("u"))
print(s.find("M"))  # Find returned Character index Number
print(s.find("s"))
print(s.find("z")) # Return -1(since its return index, -1 means not oud)
print(s.index("s"))
# print(s.index("z")) # it will return error
```

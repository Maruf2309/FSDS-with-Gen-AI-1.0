### Relational Operator

```python
print("Hello" == "Hello") # Equality
print("Hello" !="World") # not equal
print("hello" > "Hi") # Greater (in python its follow Oxford Dictinary word ordering, in Oxford dic, Hello in first then Hi- thats why True)
print("hello" < "Hi") # Greater
print("hello" < "World") # Greater
print("hello" >= "Hello") # Greater than or equal to
```

- `Hello < World` In oxford hello stays first then World- that why `False`

### Swapcase Method

This method return a string where all the upper case letter converted to lower case & vice versa

```python
t1 = "Good Morning"
t2 = "good morning"
print(t1.swapcase())
print(t2.swapcase())
```
